### importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load the dataset.


```python
df=pd.read_csv('/content/House Price India.csv')
df.head()
```

### 3. Perform the Below Visualizations

###Univariate Analysis

number of views

```python
df['number of views'].value_counts()
```

```python
sns.distplot(df['number of views'])
```

```python
sns.displot(df['number of views'])
```

```python
plt.pie(df['number of views'].value_counts(),[0,0,0.1,0.2,0.3],labels=[0,2,3,1,4],autopct='%1.1f%%',shadow=True)
plt.title('number of views')
plt.show
```

```python
sns.barplot(x=df['number of views'].value_counts().index,y=df['number of views'].value_counts())
```
